# **Problem Statement**

Create a command-line to-do application that allows a user to manage tasks using text commands. The application should run continuously until the user chooses to exit.

```python
file=open("output.txt","r")
data=file.read()
if data=="":
    print("Nothing to do !")
else:
    print(data)
tasks=[]
while True:
    print("Available commands")
    print("1.Add new task")
    print("2.Show all tasks")
    print("3.Mark task as done")
    print("4.Exit")
    command=input("Enter a command:")
    if(command.lower()=="add" or command=='1'): 
        taskNo=int(input("Enter a task Number:"))
        taskDes=input("Enter a task description:")
        status="Not done"
        tasks.append([taskNo,taskDes,status])
    elif(command.lower()=="show" or command=='2'):
        print(tasks)
    elif(command.lower()=="Mark" or command=='3'):
        utaskNo=int(input("Enter a task Number:"))
        for task in tasks:
            if task[0]==utaskNo:
                task[2]="Done"
                print("Marked as completed !")
        print(tasks)
    elif(command.lower()=="exit" or command=='4'):
        file=open("output.txt","w")
        file.write(str(tasks))
        file.close()
        print("Exited successfully!")
        break
    else:
        print("Invalid Command !!!")
```
