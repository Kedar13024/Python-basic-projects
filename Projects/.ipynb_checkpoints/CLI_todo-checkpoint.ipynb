# **Problem Statement**

Create a command-line to-do application that allows a user to manage tasks using text commands. The application should run continuously until the user chooses to exit.

```python
tasks=[]
while True:
    print("Available commands")
    print("1.Add new task")
    print("2.Show all tasks")
    print("3.Mark task as done")
    print("4.Exit")
    command=input("Enter a command:")
    if(command.lower()=="add" or command=='1'): 
        taskNo=int(input("Enter a task Number:"))
        taskDes=input("Enter a task description:")
        status="Not done"
        tasks.append([taskNo,taskDes,status])
    elif(command.lower()=="show" or command=='2'):
        print(tasks)
    elif(command.lower()=="Mark" or command=='3'):
        utaskNo=int(input("Enter a task Number:"))
        for task in tasks:
            if task[0]==utaskNo:
                task[2]="Done"
                print("Marked as completed !")
        print(tasks)
    elif(command.lower()=="exit" or command=='4'):
        print("Exited successfully!")
        break
    else:
        print("Invalid Command !!!")
```

```python
# ---------- LOAD TASKS FROM FILE ----------
tasks = []

try:
    file = open("output.txt", "r")
    for line in file:
        tasks.append(eval(line.strip()))
    file.close()
except FileNotFoundError:
    pass   # file does not exist yet


# ---------- MAIN PROGRAM ----------
while True:
    print("\nAvailable commands")
    print("1. Add new task")
    print("2. Show all tasks")
    print("3. Mark task as done")
    print("4. Exit")

    command = input("Enter a command: ")

    # ADD TASK
    if command.lower() == "add" or command == "1":
        taskNo = int(input("Enter a task number: "))
        taskDes = input("Enter a task description: ")
        status = "Not done"

        task = [taskNo, taskDes, status]
        tasks.append(task)

        print("Task added successfully!")

    # SHOW TASKS
    elif command.lower() == "show" or command == "2":
        if not tasks:
            print("Nothing to do!")
        else:
            for task in tasks:
                print(task)

    # MARK TASK AS DONE
    elif command.lower() == "mark" or command == "3":
        utaskNo = int(input("Enter task number to mark as done: "))
        found = False

        for task in tasks:
            if task[0] == utaskNo:
                task[2] = "Done"
                found = True
                print("Marked as completed!")
                break

        if not found:
            print("Task number not found!")

    # EXIT
    elif command.lower() == "exit" or command == "4":
        file = open("output.txt", "w")
        for task in tasks:
            file.write(str(task) + "\n")
        file.close()

        print("Exited successfully!")
        break

    else:
        print("Invalid Command!")
```
